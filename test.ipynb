```python
formatted_history = []

for msg in history:
    role = msg["role"]

    # extract text only (ignore tool calls for now)
    text_parts = []

    for c in msg.get("contents", []):
        if c["type"] == "text":
            text_parts.append(c["text"])

    if text_parts:
        formatted_history.append({
            "role": role,
            "content": "\n".join(text_parts)
        })
```

```python
async for update in agent.run("i need a diet plan for weight loss", stream=True):
    if update.contents and update.contents[0].type == "function_call":
        print(update.contents[0].type)
    if update.contents and update.contents[0].type == "function_result":
        print("\nCompleted\n\n")
    if update.text:
        print(update.text, end="", flush=True)
```

```python
from IPython.display import clear_output, display, Markdown
import json
import time

status = ""
text_buffer = ""
last_update = 0
skill = ""
async for update in agent.run("help me plan my diet", stream=True):
    if update.contents:
        content = update.contents[0]

        if content.type == "function_call":
            status = f"Using Skill:"
            clear_output(wait=True)   # 👈 clears cell smoothly
            display(status)
            skill += content.arguments
        elif content.type == "function_result":
            skill = __import__('json').loads(skill)["skill_name"]
            status = f"{skill} - ✅ Completed"
            clear_output(wait=True)   # 👈 clears cell smoothly
            display(status)       # 👈 shows status line
    if update.text:
        text_buffer += update.text
        # if time.time() - last_update > 0.1:  # throttle
        clear_output(wait=True)
        # display(status)
        display(Markdown(f"{status}\n\n{text_buffer}"))
        last_update = time.time()
        # text_buffer += update.text
        # clear_output(wait=True)
        # display(status) 
        # display(Markdown(text_buffer))
```
