```python
from IPython.display import clear_output, display, Markdown
from agent_framework import Message, Role
import json
import time

messages = [
]

while True:
    user_input = input("You: ")

    print("\nUser:", user_input)
    
    if user_input.lower() in ["exit", "quit"]:
        break

    # Add user message
    messages.append(Message("user", [user_input]))

    status = ""
    text_buffer = ""
    skill_buffer = ""

    # Stream response
    async for update in agent.run(messages, stream=True):
        if update.contents:
            content = update.contents[0]

            if content.type == "function_call":
                status = "Using Skill..."
                # clear_output(wait=True)
                # display(Markdown(f"{status}"))
                skill_buffer += content.arguments

            elif content.type == "function_result":
                try:
                    skill_name = json.loads(skill_buffer).get("skill_name", "Skill")
                except:
                    skill_name = "Skill"
                status = f"{skill_name} - ✅ Completed"
                # clear_output(wait=True)
                # display(Markdown(f"{status}"))

        if update.text:
            text_buffer += update.text
            # clear_output(wait=True)
            display(Markdown(f"{status}\n\n{text_buffer}"))

    # Save assistant response into memory
    messages.append(Message("assistant", [text_buffer]))
```
